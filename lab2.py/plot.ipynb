```python
import matplotlib.pyplot as plt 
from matplotlib import patches
from shape import Shape 
from circle import Circle 
from rectangle import Rectangle


c1 = Circle(0 ,0, 1)
c2 = Circle(4 ,3, 3)
rec1 = Rectangle(6, 6, 3, 3) 

fig, ax = plt.subplots(1)
figsize = (8, 6)
circle_patch1 = patches.Circle ((c1.x, c1.y), radius = c1.radius ,edgecolor = "green")
circle_patch2 = patches.Circle ((c2.x, c2.y), radius = c2.radius, edgecolor = "purple")
rectangle_patch1 = patches.Rectangle((rec1.y, rec1.x), rec1.width ,rec1.heigth ,edgecolor = "red")

ax.add_patch(circle_patch1)
ax.add_patch(circle_patch2)
ax.add_patch(rectangle_patch1)
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.grid() 
```
